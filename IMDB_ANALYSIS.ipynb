# remove/fill null values,  
#drop the unnessasary columns

```python
movies["color"].value_counts()
```

```python
#maximum movies in color so we can drop this column
movies.drop('color',axis=1,inplace=True)
```

```python
#remove movie_imdb_link the column is not required so we can remove the column
movies.drop(['movie_imdb_link'],axis=1,inplace=True)
```

```python
# We can remove the null values from the dataset where the count is less . so that we don't loose much data 

movies.dropna(axis=0,subset=['actor_1_facebook_likes','actor_1_name','actor_3_name','facenumber_in_poster',
                               'num_user_for_reviews','language','country','actor_2_facebook_likes',
                               'director_name', 'num_critic_for_reviews',
                               'duration','director_facebook_likes','actor_3_facebook_likes','actor_2_name',
                               'plot_keywords'],inplace=True)
```

# fill null values

```python
movies["content_rating"].value_counts() 
```

```python
#fill the null values by mean
movies["gross"].fillna(movies["gross"].mean(),inplace=True)
```

```python
# R most frequntly so i will fill the null values by R.
movies["content_rating"].fillna("R", inplace = True) 
```

```python
# in the budget column i fill the null values by MEAN.
movies["budget"].fillna(movies["budget"].mean(),inplace=True)
```

```python
#in the aspect_ratio fill the null values by mean. 
movies["aspect_ratio"].fillna(movies["aspect_ratio"].mean(),inplace=True)
```

```python
#now again check the null values 
movies.isnull().sum()
```

# now drop the unwanted data(duplicates) or columns

```python
movies.drop_duplicates(inplace=True)
movies.shape
```

```python
#check the net profit made by the company (Gross-Budget) 

movies["Profit"]=round(movies['gross'].sub(movies['budget'], axis = 0),2) 
```
